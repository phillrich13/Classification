## Importing data from Austin Animal Center

```python
intakes = pd.read_csv(r'/Users/philliprichardson/Metis/Module 4/Austin_Animal_Center_Intakes.csv')
outcomes = pd.read_csv(r'/Users/philliprichardson/Metis/Module 4/Austin_Animal_Center_Outcomes.csv')
```

```python
df = intakes.merge(outcomes, left_on = ['Animal ID','Animal Type'], right_on = ['Animal ID','Animal Type'])
```

```python
test = (df.groupby(['Animal ID']).DateTime_x.count() > 1).reset_index()
```

```python
mask = (test['DateTime_x'] == True)
```

```python
animal_dupes = test[mask]['Animal ID']
```
